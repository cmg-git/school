**p. 079 Inéquations de 1er degré**

```python
import numpy as np
import matplotlib.pyplot as plt

def comparison_1degree(x, param1, param2):
    a1, b1 = param1
    a2, b2 = param2
    y2 = a2 * x + b2
    y1 = a1 * x + b1

    y_diff = y1 - y2
    if a1 - a2 != 0:
        x_diff_sol = -(b1 - b2) / (a1 - a2)

    y1_label = f'$y_1 = {a1} * x + {b1}$'
    y2_label = f'$y_2 = {a2} * x + {b2}$'
    y_diff_label = '$y_{diff} = y_{1} - y_{2}$'

    fig, ax = plt.subplots()
    ax.plot(x, y1, x, y2)
    ax.set(xlabel='x',
           ylabel='y')
    ax.legend([y1_label, y2_label])

    if a1 - a2 != 0:
        ax.plot(x, y_diff, linestyle='--', color='red')
        ax.plot(x_diff_sol, 0, 'o', color='red')
        ax.legend([y1_label, y2_label, y_diff_label])
        title = f"$y_1 - y_2 = 0$ pour $x = {x_diff_sol:.2f}$"
        ax.set_title(title)

    ax.grid()
    plt.show()
```

**ex. 50** On considère l'inéquation $-4x - 40 > 60$ d'inconnue réelle $x$.

Résoudre cette inéquation.

```python
a1 = -4
b1 = -40
param1 = [a1, b1]

a2 = 0
b2 = 60
param2 = [a2, b2]

x = np.arange(start=-60, stop=60, step=0.1)
comparison_1degree(x, param1, param2)
```

**ex.55** Soit l'expression $A = 45 + 5x$ et $B = 1000 - 5x$. Comment faut-il choisir $x$ pour que le resultat de $A$ soit supérier au résultat de $B$ ?
